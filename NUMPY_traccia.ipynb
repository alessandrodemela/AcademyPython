````python
def shift(arr):   #funzione di trasformazione di una lista nella lista shiftata destra di 1
    return np.concatenate((arr[-1:],arr[0:-1]))
````

````python
shift([1,2,3,4])
````

````python
#Esempio...come implementare un cubo di rubik

class Cubo_Rubik:
    
    def shift(self,arr):   #funzione di trasformazione di una lista nella lista shiftata destra di 1
        import numpy as np
        return np.concatenate((arr[-1:],arr[0:-1]))
    
    def __init__(self):
        import numpy as np
        mat=np.ones((3,3),dtype=int)
        self.facce=np.array([mat*i for i in range(6)]) #creo le facce del cubo
    
    def test_ruota(self): #metodo di test che ruota la prima colonna della prima faccia
        self.facce[:4,:,0]=self.shift(self.facce[:4,:,0])
         
        
````
